**1. For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.**

- **(a) The sample size n is extremely large, and the number of predictors p is small.**

A more flexible approach will fit the data closer and with the large sample size a better fit would be obtained than with an infledible approach.

- **(b) The number of predictors p is extremely large, and the number of observations n is small.**

An inflexible method would be better so as to not overfit the small number of observations.

An inflexible method would avoid fitting the noise in the error terms and decrease the variance.

**3. We now revisit the bias-variance decomposition.**
- **(a) Provide a sketch of typical (squared) bias, variance, training error, test error, and Bayes (or rreducible) error curves, on a single plot, as we go from less flexible statistical learning methods towards more flexible approaches. The x-axis should represent the amount of flexibility in the method, and the y-axis should represent the values for each curve. There should be five curves. Make sure to label each one**

```python
# Plot
x = np.linspace(0,10,10)

v_err = [10, 8.3, 7.0, 5.9, 5.2, 5.0, 5.2, 5.9, 7.0, 8.3, 10]
bayes = 4 * np.ones(10)
var = [0.1, 0.7, 1.2, 1.7, 2.3, 3.1, 4.2, 5.3, 6.9, 8.8]
bias = [8.5, 5.4, 3.3, 2.2, 1.9, 1.7, 1.4, 1.2, 1.0, 0.9]
t_err = [9.5, 7.3, 5.7, 4.4, 3.5, 2.7, 1.9, 1.1, 0.5, 0.0]

fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])

ax.set_xlabel('Flexibility')
ax.set_ylabel('Curve Values')

ax.plot(np.linspace(0,10,11), v_err, label='Validation (Test) Error', lw=3)
ax.plot(x, bayes, '--', label='Bayes error', lw=3)
ax.plot(x, var, label='Variance', lw=3)
ax.plot(x, bias, label='Bias', lw=3)
ax.plot(x, t_err, label='Training Error', lw=3)

ax.set_xlim([-1,11])
ax.set_ylim([-1,11])

ax.legend(loc=(1.1,0.6))
plt.show()
```

- **(b) Explain why each of the five curves has the shape displayed in part (a).**

When a model with very low flexibility, such as a linear model, has its predictors changed, the resulting model will vary very little. This means lower flexibility translates to lower variance. The opposite is true for bias, and this is known as the bias-variance trade-off. Training error can be minimized by increasing flexibility, but Bayes error (or irreducible error) will make it so that more flexibility beyond a certain point will continue to decrease training error, but test error will begin to rise.

**6. Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classification (as opposed to a non-parametric approach)? What are its disadvantages?**

A *parametric approach* reduces the problem of estimating $f$ down to one of the estimating a set of parameters because it assumes a form for $f$.  
A *non-parametric approach* does not assume a form for $f$ and so requires a very large number of observations to accurately estimate $f$.

The *advantages* of a parametric approach to regression or classification are the simplifying of modeling $f$ to a few parameters and not as many observations are required compared to a non-parametric approach.  
The *disadvantages* of a parametric approach to regression or classification are a potential to inaccurately estimate $f$ if the form of $f$ is werong or to possibly overfit the observations if more flexible models are used.
